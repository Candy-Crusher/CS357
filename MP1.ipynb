```python
# MP 1.1
import numpy as np
def cardNameToInt(card):
    # card: string following the "Rank Suit format"
    # write code to convert hand to an integer 0-51 as described above
    ranks = ['2','3','4','5','6','7','8','9','T','J','Q','K','A']
    rank_dict = {}
    suits = ['c','d','h','s']
    suit_dict = {}
    for i in range(len(ranks)):
        rank_dict[ranks[i]] = 4*i
    for j in range(len(suits)):
        suit_dict[suits[j]] = j
    return rank_dict[card[0]]+suit_dict[card[1]]
start_cards = np.zeros(len(starting_hand), int)
for i in range(len(starting_hand)):
    start_cards[i] = cardNameToInt(starting_hand[i])
```

```python
# MP 1.2
import numpy as np
import random
def generatePlayersCards(n_players,available_deck):
    # do things here
    players_cards = np.zeros((n_players, 2), int)
    for p in range(n_players):
        for i in range(2):
            while True:
                new = random.randint(0, 51)
                if new not in players_cards and new in available_deck:
                    players_cards[p][i] = new
                    available_deck = np.delete(available_deck, np.where(available_deck == new))
                    break
       
    return(players_cards,available_deck)

def generateDealerCards(available_deck):
    dealer_cards = np.zeros(5, int)
    for p in range(5):
        while True:
            new = random.randint(0, 51)
            if new not in dealer_cards and new in available_deck:
                dealer_cards[p] = new
                available_deck = np.delete(available_deck, np.where(available_deck == new))
                break
    return(dealer_cards,available_deck)
```

```python
# MP 1.5
import numpy as np
def monteCarloSimulation(M,N,n_players,starting_hand):
    win_prob = np.zeros(M)
    for m in range(M):
        win = 0
        for game in range(N):
            start_cards = np.zeros(2, int)
            available_deck = np.arange(52)
            for i in range(2):
                start_cards[i] = cardNameToInt(starting_hand[i])
                available_deck = np.delete(available_deck, np.where(available_deck == start_cards[i]))
            players_cards,available_deck = generatePlayersCards(n_players,available_deck)
            dealer_cards,available_deck = generateDealerCards(available_deck)
            game_result = whoWin(start_cards, players_cards, dealer_cards)
            if game_result[0] == 1:
                win += 1
        win_prob[m] = win / N
    return win_prob
```
